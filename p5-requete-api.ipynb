## Notebook des requetes de StackAPI

```python
# Importation des librairies

from stackapi import StackAPI
import pandas as pd
import numpy as np
```

```python
# Charger l'API et definir le nombre de resultat a obtenir

SITE = StackAPI('stackoverflow')
SITE.page_size = 2      # Nombre de resultat * 'max_pages'
SITE.max_pages = 25     # Nombre de pages 'max_pages'
```

```python
# Recuperer les 50 questions ayant le tag 'python' et un vote minimum de 50 depuis le 01/01/2020 jusqu'au 01/01/2024

questions = SITE.fetch('questions', tagged='python', min=50, sort='votes', fromdate=1577836800, todate=1704067200)
```

```python
# Creation du DataFrame et selection des variables 'title', 'tags', 'score', 'creation_date'

dataset = pd.DataFrame(questions['items'])
dataset = dataset[['title', 'tags', 'score', 'creation_date']]
dataset['creation_date'] = pd.to_datetime(dataset['creation_date'], unit='s')
```
